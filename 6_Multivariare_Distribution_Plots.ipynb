```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('https://raw.githubusercontent.com/datagrad/Learning-Python/main/Marks%20Data.csv')
df.head()
```

```
fig, ax = plt.subplots(figsize=(10.5, 7.2))

sns.scatterplot(x='Total', y='Science', data=df, color='#F15B2A',hue='Sports', ax=ax)
```

```
sns.jointplot(x='Total', y='Math', data=df, color='#F15B2A',hue='Year', kind='kde')
```

```
fig = plt.figure(figsize=(21, 14.4))
sns.kdeplot(df['Total'], df['Math'])
```

```
from mpl_toolkits.mplot3d import Axes3D
```

```
fig = plt.figure(figsize=(10.5, 7.2))
ax = fig.gca(projection='3d')

ax.plot_trisurf(df['Total'], df['Math'], df['Year'], color='#F15B2A')
```

```
fig = plt.figure(figsize=(15, 9))
ax = fig.add_subplot(111, projection='3d')

ax.bar3d(df['Total'], df['Math'], np.zeros(len(df['Science'])), 1, 1, df['Science'], shade=True, color='#F15B2A') 
```

```
fig, ax = plt.subplots(figsize=(10.5, 7.2))

sns.boxplot(x='Year', y='Science', hue='Sports', data=df, ax=ax)
```

```
fig, ax = plt.subplots(figsize=(10.5, 7.2)) 

sns.violinplot(x='Total', y='Attandence', hue='Acting', data=df, ax=ax)
```
